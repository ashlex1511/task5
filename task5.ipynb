```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
aa=pd.read_csv(r"C:\Users\Aashi\Downloads\heart.csv")
```

```python
sns.set(style="whitegrid", palette="muted")
```

```python
palette= sns.color_palette("Set2")
```

```python
plt.figure(figsize=(12,6))
sns.histplot(aa['age'], bins=30, kde=True, color=palette[0])
plt.title('Distribution of Age', fontsize=17)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x= 'sex', y='chol', data=aa, palette="Set2")
plt.title('Cholestrol levels by Sex', fontsize=17)
plt.xlabel('Sex(0= Female, 1=Male)', fontsize=11)
plt.ylabel('Cholestrol', fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(12,6))
corr_mtrix=aa.corr()
sns.heatmap(corr_mtrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matrix', fontsize=17)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(12,6))
mean1=aa.groupby('target')['thalach'].mean().reset_index()
sns.barplot(x='target', y='thalach', data=mean1, palette="Set2")
plt.title('Average Max Heart Rate by Heart Disease Presence', fontsize=17)
plt.xlabel('Heart Disease (0=No, 1=Yes)', fontsize=11)
plt.ylabel('Avg Max Heart Rate', fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='age', y='thalach', data=aa, hue='target', palette="Set2", s=100)
plt.title('Age Vs max heart rate achieved', fontsize=17)
plt.xlabel('AGe', fontsize=11)
plt.ylabel('Max heart rate achieved', fontsize=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
sns.despine()
plt.show()
```
